# Identifying Missing Data

```python
# number of rows missing in each column
df.isnull().sum().sort_values(ascending = False)
```

```python
# % of rows missing in each column
for column in df.columns:
    percentage = df[column].isnull().mean()
    print(f'{column}: {round(percentage*100, 2)}%')
```

# Dealing with missing data

```python
# Remove a column or row with .drop, .dropna or .isnull
```

```python
#drop colunm
df.drop('description', axis=1)
```

```python
#drop row
df.drop('director', axis=1)
# drop row
no_director = df[df['director'].isnull()].index
df.drop(no_director, axis=0)
# drop row
# no_duration = df[df['duration'].isnull()].index
# df.drop(no_duration, axis=0, inplace=True)

# ~ + .isnull()
df[~df['director'].isnull()]

# dropna()
df.dropna(subset=['director'])
```

```python
# Replace it by the mean, median (numeric data) or mode (categorical data)
df[df['rating'].isnull()]
mode = ''.join(df['rating'].mode())
df['rating'].fillna(mode, inplace=True)
```

# Remove blank spaces with strip(), lstrip(), and rstrip()

```python
movie_title = " Titanic   "
```

```python
# trim leading space with lstrip()
print(movie_title.lstrip())
```

```python
# trim leading space with rstrip()
print(movie_title.rstrip())
```

```python
# trim leading space with strip()
print(movie_title.strip())
```

```python
# get rid of leading and trailing white spaces with the strip() method
df['title'] = df['title'].str.strip()

# get rid of leading and trailing white spaces with the apply method
df['title'].apply(lambda x:x.strip())
```

```python
# get rid of leading and trailing white spaces with the strip() method
df['title'] = df['title'].str.rstrip()

# get rid of leading and trailing white spaces with the apply method
df['title'].apply(lambda x:x.rstrip())
```
